```python
%matplotlib inline

#import gdal
#import itertools  
import matplotlib.pyplot as plt
import numpy as np
import numpy.ma as ma
import pandas as pd
#import shapefile as shp
from rasterstats import zonal_stats
from spectral import *
import spectral.io.envi as envi
```

###Read, get files

```python
#hyperspectral image
img1 = envi.open('/media/spp/SEBAS_HDD/RSProject/2014Hyperspectral/NIS1_20140507_144927_atmcor_subset1.hdr')
img2 = envi.open('/media/spp/SEBAS_HDD/RSProject/2014Hyperspectral/NIS1_20140507_150703_atmcor_subset1.hdr')
img3 = envi.open('/media/spp/SEBAS_HDD/RSProject/2014Hyperspectral/NIS1_20140507_152342_atmcor_subset1.hdr')
img4 = envi.open('/media/spp/SEBAS_HDD/RSProject/2014Hyperspectral/NIS1_20140507_152342_atmcor_subset2.hdr')


#good band list from Sarah
goodbands = pd.read_csv('/media/spp/SEBAS_HDD/RSProject/2014Hyperspectral/badbandlist2014.csv')['band2014'].values.tolist()
goodbands  = filter(None, goodbands)

#srcImage = gdal.Open('/media/spp/SEBAS_HDD/RSProject/2014Hyperspectral/NIS1_20140507_144927_atmcor_subset1.bsq')
#srcImage.GetGeoTransform()
geotransform1 = [403772.0, 1.0, 0.0, 3286914.0, 0.0, -1.0] #_144927_atmcor_subset1
geotransform2 = [403264.0, 1.0, 0.0, 3285960.0, 0.0, -1.0] #_150703_atmcor_subset1
geotransform3 = [402287.0, 1.0, 0.0, 3285126.0, 0.0, -1.0] #_152342_atmcor_subset1
geotransform4 = [402241.0, 1.0, 0.0, 3283771.0, 0.0, -1.0] #_152342_atmcor_subset2


fid_list = pd.read_csv('/media/spp/SEBAS_HDD/RSProject/Crown Shapefiles/osbs_crowns_14april2015_forSebas/osbs_crowns_working.csv').fid.values.tolist()
#crowns.bbox
```

###Try to extract pixel band values and their species and tree id

```python
#columns = np.arange(int(img.metadata['bands']))   #, 'std']))
df = pd.DataFrame()
#df = df.fillna(0) # with 0s rather than NaNs

#fid_range = range(74)   #If we are using all of the polygons
fid_range = fid_list 
nfid = []


for b in goodbands:  #I'm creating a long column
    band1 = img1.read_band(b-1)  #python starts at 0
    band2 = img2.read_band(b-1)  #python starts a
    band3 = img3.read_band(b-1)  #python starts a
    band4 = img3.read_band(b-1)  #python starts a
    
    #To create zonal stats for eacf of the fid shapes. The raster_out value makes it return the mini_raster to
    #get the individual pixel values.
    stats1 = zonal_stats("/media/spp/SEBAS_HDD/RSProject/Crown Shapefiles/osbs_crowns_14april2015_forSebas/osbs_crowns_working.shp",
                band1, 
                transform = geotransform1,
                nodata_value = 0,
                raster_out = True)
    stats2 = zonal_stats("/media/spp/SEBAS_HDD/RSProject/Crown Shapefiles/osbs_crowns_14april2015_forSebas/osbs_crowns_working.shp",
                band2, 
                transform = geotransform2,
                nodata_value = 0,
                raster_out = True)
    stats3 = zonal_stats("/media/spp/SEBAS_HDD/RSProject/Crown Shapefiles/osbs_crowns_14april2015_forSebas/osbs_crowns_working.shp",
                band3, 
                transform = geotransform3,
                nodata_value = 0,
                raster_out = True)
    stats4 = zonal_stats("/media/spp/SEBAS_HDD/RSProject/Crown Shapefiles/osbs_crowns_14april2015_forSebas/osbs_crowns_working.shp",
                band4, 
                transform = geotransform4,
                nodata_value = 0,
                raster_out = True)

    #The polygons are not all in one raster file. This code below finds the raster that contains the polygon
    #and adds the values to the overall table. I don't like using ifs, but with four files may not be that bad
    bandvalues = np.array([])   #a list of the the band value for all the pixels.
    for fid in fid_range:    #each item in stats is the values for each fid
        a = ma.getdata(ma.getdata(stats1[fid].get('mini_raster'))).flatten() #Get 1-D list of values
        a = a [a != 0]   #remove zeros
        if len(a) == 1:
            a = ma.getdata(ma.getdata(stats2[fid].get('mini_raster'))).flatten() #Get 1-D list of values
            a = a [a != 0]  #remove zeros and change scale. Some numbers are in the 100000 scale
            if len(a) == 1:
                a = ma.getdata(ma.getdata(stats3[fid].get('mini_raster'))).flatten() #Get 1-D list of values
                a = a [a != 0] # #remove zeros
                if len(a) == 1:
                    a = ma.getdata(ma.getdata(stats4[fid].get('mini_raster'))).flatten() #Get 1-D list of values
                    a = a [a != 0] #    #remove zeros remove zeros and change scale. Some numbers are in the 100000 scale
        
        #Algunos valores regresan multiplicados por 1000. No tengo idea porqué. El código abajo corrige esto.
        if a[0] > 10000:
            a = a/1000
        bandvalues = np.append(bandvalues, a) #append fid valuest to the band list
        
        #To create an array of the fid numbers for each pixel in the table. Only needed once (hopefully)
        if b == 2:  #some band that will be entered into the table 
            sfid = len(a) * [fid]
            nfid = np.append(nfid, sfid)
    
    #Algunas bandas regresan un número diferente de pixeles totales. Mejor me salto esas bandas.
    #print len(bandvalues)
    if len(bandvalues) == 4858 : #5283 if we are using all polygons
        df['band' + str(b)] = bandvalues   #adds a new column with band value for all fid
df['fid'] = nfid    #Añade una columna con el número de shapefile

#Export dataframe to csv file
df.to_csv('/media/spp/SEBAS_HDD/RSProject/Results/reflectance_per_tree2014_alltrees.csv')
```

#Classification

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
import seaborn as sns
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from skll.metrics import kappa
```

###Reading Files

```python
#df = pd.read_csv('/media/spp/SEBAS_HDD/RSProject/Results/reflectance_per_tree2014_8trees.csv')
df = pd.read_csv('/media/spp/SEBAS_HDD/RSProject/Results/reflectance_per_tree2014_alltrees.csv')

bandlist = pd.read_csv('/media/spp/SEBAS_HDD/RSProject/2014Hyperspectral/badbandlist2014.csv')
goodband_num = np.array(bandlist['goodband_num'].values.tolist()) #goodband list with zeroes of badbands
goodband_num = [x for x in goodband_num if x != 'nan'] #eliminating nan values

#Add the species for each line. It requires a csv file with the fid -> spp relation
dffid_spp = pd.read_csv('/media/spp/SEBAS_HDD/RSProject/Crown Shapefiles/osbs_crowns_14april2015_forSebas/osbs_crowns_working.csv')
fid_spp = dict(zip(dffid_spp.fid, dffid_spp.Species)) #dictionary to map later. Faster for new columns instead of a for loop
df['spp'] = df.fid.map(fid_spp.get) #New column with spp
```

###Parameters

```python
#Training fid and validation
#training_num =  40 # 74 total 
#rand_items = random.sample(df.fid.unique(), len(df.fid.unique()) )
#training_fid = rand_items[:training_num ]
#validation_fid = rand_items[training_num: ]

# choose from all species
training_fid = np.sort([31,
32,
20,
45,
64,
21,
43,
23,
1,
4,
6,
14,
47,
51,
8,
11,
27,
55,
70,
22,
50,
53,
49,
56,
0,
46,
61,
])

validation_fid = np.sort([73,
26,
58,
59,
66,
68,
67,
24,
5,
9,
10,
15,
48,
62,
7,
12,
17,
18,
52,
54,
60,
69,
16,
19,
25,
28,
30,
65,
63,
57,
2,
3,
71, 72])
```

```python
#dataframe and list of X and Y training values 
Y_train =  list(df[df.fid.isin( training_fid )].spp)  #list of trainign Y values (species)
df_train = df[df.fid.isin( training_fid )][goodband_num] #dataframe of and listX training values
X_train = map(list, df_train.values)  #making it a list to input to classifier (#training pixels times # bands)

#dataframe and list of X and Y validation values
Y_val_true = list(df[df.fid.isin( validation_fid )].spp) #list of validation Y values (species)
df_val = df[df.fid.isin( validation_fid )][goodband_num] #dataframe of and listX validation values
X_val = map(list, df_val.values) #making it a list to input to classifier (#validation pixels times # bands)

#labels for confusion matrix
spp_labels = np.sort(df.spp.unique())   #all species in the population. Some species may have one or two trees
label_true = ['true_' + s for s in spp_labels ] 
label_pred = ['pred_' + s for s in spp_labels ] 
```

####Helper functions

```python
#Function to get a dataframe on the pixel species prediction counts
def count(predictions):
    counts = pd.DataFrame(predictions.value_counts())  #Count number of species prediction per pixel per fid (crown)
    counts.reset_index(inplace=True) #Delete index
    counts = counts.pivot(index='fid', columns='level_1', values=0) #reshaping tabl
    return counts

#Function to return a dataframe of the confusion matrix
def df_conf(y_true, y_pred, labels = spp_labels, index = label_true, columns = label_pred):
    df = pd.DataFrame(confusion_matrix(y_true,
                                       y_pred,
                                       labels),
                            index,
                            columns)
    
    df.index = labels
    df.columns  = labels
    df.index.name = 'True'
    df.columns.name = 'Predicted'
    return df
```

### SVM and Random Forests

```python
##########SVM
#svm rbf
SVM1 = svm.SVC(kernel = 'rbf')
#Fitting the svm object
SVM1.fit(X = X_train, y = Y_train)
#Prediction of values with svm. It is predicting only the X values of the validation set
SVM_Y_pred1 = SVM1.predict(X = X_val)

#svm linear
SVM2 = svm.SVC(kernel = 'linear')
#Fitting the svm object
SVM2.fit(X = X_train, y = Y_train)
#Prediction of values with svm. It is predicting only the X values of the validation set
SVM_Y_pred2 = SVM2.predict(X = X_val)

#svm polynomal
SVM3 = svm.SVC(kernel = 'poly')
#Fitting the svm object
SVM3.fit(X = X_train, y = Y_train)
#Prediction of values with svm. It is predicting only the X values of the validation set
SVM_Y_pred3 = SVM3.predict(X = X_val)


##########Random Forests
RF1 = RandomForestClassifier(n_estimators=10)
#Fitting the rf object
RF1.fit(X = X_train, y = Y_train)
#Prediction of values with rf. It is predicting only the X values of the validation set
RF_Y_pred = RF1.predict(X = X_val)
```

###Analysis by pixel

```python
#Confusion matrix tables 
SVMconf1 = df_conf(y_true = Y_val_true, y_pred = SVM_Y_pred1)
SVMconf2 = df_conf(y_true = Y_val_true, y_pred = SVM_Y_pred2)
SVMconf3 = df_conf(y_true = Y_val_true, y_pred = SVM_Y_pred3)
RFconf = df_conf(y_true = Y_val_true, y_pred = RF_Y_pred)

###Normalizing results
true_spp_num  = pd.DataFrame(RFconf.sum(axis = 1)+.0)  #number of pixels per species in validation sample. Added .0 for float computation
for i in range(len(true_spp_num)):
    RFconf.loc[[i]] = RFconf.loc[[i]] / true_spp_num[0][i]
    SVMconf1.loc[[i]] = SVMconf1.loc[[i]] / true_spp_num[0][i]    
    SVMconf2.loc[[i]] = SVMconf2.loc[[i]] / true_spp_num[0][i]    
    SVMconf3.loc[[i]] = SVMconf3.loc[[i]] / true_spp_num[0][i]
```

### Analysis by Crown

```python
#Table to store results
predictions = pd.DataFrame()
predictions['SVM_Y_pred1'] = SVM_Y_pred1 #Adding all pixel predictions for SVM
predictions['SVM_Y_pred2'] = SVM_Y_pred2 #Adding all pixel predictions for SVM
predictions['SVM_Y_pred3'] = SVM_Y_pred3 #Adding all pixel predictions for SVM
predictions['RF_Y_pred'] = RF_Y_pred #Adding all pixel predictions for RF
predictions['fid'] = np.array(df[df.fid.isin(validation_fid)].fid) #Adding FID for for all pixels
predictions['Y_val_true'] = Y_val_true #Adding all real pixel values
predictions = predictions.groupby(by = 'fid') #To display results by FID number (crown)

# count of species prediction pixels in each crown
SVMcounts1 = count(predictions.SVM_Y_pred1)
SVMcounts2 = count(predictions.SVM_Y_pred2)
SVMcounts3 = count(predictions.SVM_Y_pred3)
RFcounts = count(predictions.RF_Y_pred)

#Create a crown species validation  list
Y_val_true_crown = list(SVMcounts1.index.map(fid_spp.get))

SVM_Y_pred_crown1 = [] #These lists will be filled with the majority prediction
SVM_Y_pred_crown2 = [] #These lists will be filled with the majority prediction
SVM_Y_pred_crown3 = [] #These lists will be filled with the majority prediction
RF_Y_pred_crown = []

for fid in SVMcounts1.index:
    #Will grab the species with the majority
    SVM_Y_pred_crown1.append(SVMcounts1.loc[fid][SVMcounts1.loc[fid] == SVMcounts1.loc[fid].max()].index[0])
    SVM_Y_pred_crown2.append(SVMcounts2.loc[fid][SVMcounts2.loc[fid] == SVMcounts2.loc[fid].max()].index[0])
    SVM_Y_pred_crown3.append(SVMcounts3.loc[fid][SVMcounts3.loc[fid] == SVMcounts3.loc[fid].max()].index[0])
    RF_Y_pred_crown.append(RFcounts.loc[fid][RFcounts.loc[fid] == RFcounts.loc[fid].max()].index[0])

#Confusion matrix tables 
SVMconf_crown1 = df_conf(y_true = Y_val_true_crown, y_pred = SVM_Y_pred_crown1)
SVMconf_crown2 = df_conf(y_true = Y_val_true_crown, y_pred = SVM_Y_pred_crown2)
SVMconf_crown3 = df_conf(y_true = Y_val_true_crown, y_pred = SVM_Y_pred_crown3)
RFconf_crown = df_conf(y_true = Y_val_true_crown, y_pred = RF_Y_pred_crown)
    
###Normalizing results
true_spp_num_crown  = pd.DataFrame(RFconf_crown.sum(axis = 1)+.0) #number of crowns per species in validation sample. Added .0 for float computation
for i in range(len(true_spp_num)):
    RFconf_crown.loc[[i]] = RFconf_crown.loc[[i]] / true_spp_num_crown[0][i]
    SVMconf_crown1.loc[[i]] = SVMconf_crown1.loc[[i]] / true_spp_num_crown[0][i]
    SVMconf_crown2.loc[[i]] = SVMconf_crown2.loc[[i]] / true_spp_num_crown[0][i]
    SVMconf_crown3.loc[[i]] = SVMconf_crown3.loc[[i]] / true_spp_num_crown[0][i]
```
